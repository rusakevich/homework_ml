```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from pymystem3 import Mystem
import pickle
```

```python
# Чтение данных
with open('duma/data_dataframe.pickle', 'rb') as dep:
    data = pickle.load(dep)
```

```python
# Функция для удаления фио из стенограмм
def fio(x):
    a = x['name'].lower().split()
    f = a[0]
    i = a[1][0]
    o = a[2][0]
    a = '{} {} {} '.format(f, i, o)
    return re.sub(a, "", x['_lines'])
```

```python
data['_lines'] = data['_lines'].apply(lambda x: re.sub(r'\s+', ' ', ' '.join(x)))
data['_lines'] = data['_lines'].apply(lambda x: re.sub(r'[^\w\s]', '', x.lower()))
data['_lines'] = data['_lines'].apply(lambda x: x.strip())
data['_lines'] = data.apply(fio, axis=1)
```

```python
# Создание объекта лематизатора, загрузка стоп-слов и русского словаря
lem = Mystem()
nltk.download('stopwords')
stop = stopwords.words('russian')
```

```python
# Функция лематизации
def prep(x):
    return [word.rstrip() for word in (''.join(lem.lemmatize(x))).split(' ') if (word not in stop and word != '')]
```

```python
%%time
# Лематизация
data['lines'] = data['_lines'].apply(prep)
```
